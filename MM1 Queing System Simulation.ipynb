```python
# export sim results to excel
# editable mu and lambda
import pandas as pd
import math
import random
from operator import itemgetter

mymu=0.1
mylambda=0.125
rho=mylambda/mymu
n=100000
iat=[]
st=[]
for i in range(n):
    iat.append(-math.log(1-random.uniform(0,1))/mylambda)
    st.append(-math.log(1-random.uniform(0,1))/mymu)
at=[]
at.append(0)
st_start=[]
st_start.append(0)
st_end=[]
st_end.append(st[0])
wq=[]
w=[]
idle=[]
idle.append(0)
for i in range(1,n):
    at.append(at[i-1]+iat[i-1])
    st_start.append(max(at[i],st_end[i-1]))
    st_end.append(st_start[i]+st[i])
    idle.append(st_start[i]-st_end[i-1])
wait=[]
for i in range(n):
    wq.append(st_start[i]-at[i])
    w.append(wq[i]+1/mymu)
    if at[i]==st_start[i]:
        wait.append(False)
    else:
        wait.append(True)
temp=[]
entList=[]
for i in range(n):
    temp.append(at[i])
    temp.append('a') #arrival
    entList.append(temp)
    temp=[]
    temp.append(st_start[i])
    temp.append('s') #start service
    entList.append(temp)
    temp=[]
    temp.append(st_end[i])
    temp.append('d') #departure
    entList.append(temp)
    temp=[]
entList=sorted(entList,key=itemgetter(0))
dt=[]
for i in range(len(entList)-1):
    dt.append(entList[i+1][0]-entList[i][0])
dt.append(0)
ent_L=[]
ent_Lq=[]
ent_L.append(1)
ent_Lq.append(1)
is_at=[] #is arrival
is_st=[] #is start service
is_dp=[] #is departure
for i in range(len(entList)):
    if entList[i][1]=='a':
        is_at.append(1)
    else:
        is_at.append(0)
    if entList[i][1]=='s':
        is_st.append(1)
    else:
        is_st.append(0)
    if entList[i][1]=='d':
        is_dp.append(1)
    else:
        is_dp.append(0)
for i in range(1,len(entList)):
    ent_L.append(ent_L[i-1]+is_at[i]-is_dp[i])
    ent_Lq.append(ent_Lq[i-1]+is_at[i]-is_st[i])
dt_L=[]
dt_Lq=[]
for i in range(len(entList)):
    dt_L.append(dt[i]*ent_L[i])
    dt_Lq.append(dt[i]*ent_Lq[i])
L=sum(dt_L)/at[n-1]
Lq=sum(dt_Lq)/at[n-1]
print("L=",L)
print("Lq=",Lq)
print("wq=",sum(wq)/len(wq))
print("w=",sum(wq)/len(wq)+1/mymu)
d={'Arrival Time':at,'Inter-arrival Time':iat,
   'Starting Service Time':st_start,
  'Service Time':st,
   'Finishing Service Time':st_end,
   'Idle Time':idle,
   'Wait':wait,
   'Average Waiting Time in Queue':wq,
   'Average Waiting Time in System':w}
df=pd.DataFrame(data=d)
df
```
